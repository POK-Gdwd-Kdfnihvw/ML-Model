### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

from plotly.offline import iplot
import plotly as py
import plotly.tools as tls

import pickle
```

### Data Preprocessing

#### Dataset Description<br>
**Age**: Age in Years<br>
**Sex**: Sex (1=Male, 0=Female)<br>
**Cp**: Chest Pain Type (0=Typical Angina, 1=Atypical Angina, 2=Non Anginal Pain, 3=Asymptomatic)<br>
**Trestbps**: Resting Blood Pressure in mmHg<br>
**Chol**: Serum Cholesterol in mg/dL<br>
**fbs**: Fasting Blood Sugar > 120 mg/dL (1=True, 0=False)<br>
**Restecg**: Resting Electrocardiographic Results ( 0=Normal, 1=Having ST-T Wave Abnormality, 2=Probable or Definite Left Ventricular Hypertrophy)<br>
**Thalach**: Maximum Heart Rate Achieved<br>
**Exang**: Exercise Induced Angina (1=Yes, 0=No)<br>
**Oldpeak**: ST Depression Induced by Exercise Relative to Rest<br>
**Slope**: Slope of the Peak Exercise ST Segment (0=Upsloping, 1=Flat, 2=Downsloping)<br>
**Ca**: Number of Major Vessels Colored by Fluorosopy<br>
**Thal**: Thalassemia (3=Normal, 6=Fixed Defect, 7=Reversable Defect)<br>
**Target**: Heart Disease (1=No, 2=Yes)<br><br>

```python
heart_data = pd.read_csv('data.csv')
heart_data.head()
```

#### Dataset Column Alteration

```python
heart_data.rename(columns = {'age':'Age',
                             'sex':'Sex',
                             'cp':'Chest Pain Type',
                             'trestbps':'Resting Blood Pressure in mmHg',
                             'chol':'Serum Cholesterol in mg/dL',
                             'fbs':'Fasting Blood Sugar > 120 mg/dL',
                             'restecg':'Resting Electrocardiographic Results',
                             'thalach':'Maximum Heart Rate Achieved',
                             'exang':'Exercise Induced Angina',
                             'oldpeak':'ST Depression Induced by Exercise Relative to Rest',
                             'slope':'Slope',
                             'ca':'Major Vessels Colored by Fluorosopy',
                             'thal':'Thalassemia',
                             'target':'Target'
                            }, inplace = True)
```

```python
heart_data.head()
```

```python
heart_data.isnull().sum()
```

### Data Visualization

### Implementation of ML Model

#### Setting the Features and label

```python
label = heart_data['Target']
features = heart_data.drop(['Target'], axis = 1)
```

#### Splitting the Data

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size = 0.4, random_state = 0)
```

#### Training and Evaluating Model

```python
def train_and_evaluate_model(model, X_train, y_train, X_test, y_test):
    results = {}
    
    model.fit(X_train, y_train)
    y_predicted = model.predict(X_test)
    
    results['classification_report'] = classification_report(y_test, y_predicted)
        
    return results
```

#### Initializing the Model

```python
decision_tree_classifier = DecisionTreeClassifier(random_state = 1)
random_forest_classifier = RandomForestClassifier(random_state = 1)
gradient_boosting_classifier = GradientBoostingClassifier(random_state = 1)

results = {}

for cls in [decision_tree_classifier, random_forest_classifier, gradient_boosting_classifier]:
    cls_name = cls.__class__.__name__
    results[cls_name] = {}
    results[cls_name] = train_and_evaluate_model(cls, X_train, y_train, X_test, y_test)
```

#### Printing the Results

```python
for result in results:
    print ("Classification Report :", result)
    print()
    for i in results[result]:
        print(results[result][i])
        print()
    print()
```

### Finalizing the ML Model

```python
decision_tree_classifier = DecisionTreeClassifier(random_state = 1)
random_forest_classifier = RandomForestClassifier(random_state = 1)
gradient_boosting_classifier = GradientBoostingClassifier(random_state = 1)

results = {}

for cls in [decision_tree_classifier, random_forest_classifier, gradient_boosting_classifier]:
    cls_name = cls.__class__.__name__
    results[cls_name] = {}
    results[cls_name] = train_and_evaluate_model(cls, X_train, y_train, X_test, y_test)
```

#### Selecting Random Forest from the Above Result

```python
important_features = random_forest_classifier.feature_importances_

for x,y in enumerate(important_features):
    print('Feature: %s, Score: %.5f' % (features.columns[x], y))

plt.bar([x for x in range(len(important_features))], important_features)
plt.show()
```

#### Saving the Model in .pkl

```python
with open('model_heart_disease.pkl', 'wb') as file:
    pickle.dump(random_forest_classifier, file)
```
